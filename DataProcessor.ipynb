```python
import pandas
import os
```

```python
data_dir='/home/vaisakh/developer/modeling/tmp/extract/'

in_data_profcsv  = os.path.join(data_dir,'IdleSleep-perf-1.prof.csv')
in_data_power    = os.path.join(data_dir,'IdleSleep-perf-1.powdata')
in_data_polldata = os.path.join(data_dir,'IdleSleep-perf-1.polldata')
```

```python
prof_df = pandas.read_csv(in_data_profcsv)

power_df = pandas.read_csv(in_data_power)
power_df.drop(['sm_mstime', 'localtime',
               'ps_ippwr-volts_mV', 'ps_ippwr-ampere_mA', 'ps_ippwr-watt_mW', 'ps_ippwr-status_b',
               'dev_ippwr-ch0-volts_mV', 'dev_ippwr-ch0-ampere_mA',	'dev_ippwr-ch0-watt_mW', 'dev_ippwr-ch0-status_b', 'dev_ippwr-ch0-interrupts',
               	'dev_ippwr-ch1-status_b', 'dev_ippwr-ch1-interrupts',
                'crc8-2sc','crc8-xor'],
              axis=1, 
              inplace=True)
poll_df = pandas.read_csv(in_data_polldata)
poll_df.drop( ['ts_local', 
               'stat_cpuid', 'stat_user', 'stat_nice', 'stat_system', 'stat_idle', 'stat_iowait',
               'stat_irq', 'stat_softirq', 'stat_steal', 'stat_guest', 'stat_guest_nice'
               ],
            axis=1, 
            inplace=True)

print ('Perf/Profile Data:')
display(prof_df)

print ('Power data:')
display(power_df)

print('Polled data')
display(power_df)
```

```python
prof_df["utctime"] = pandas.to_datetime(prof_df['utctime'])
prof_df.index = prof_df["utctime"]

power_df["utctime"] = pandas.to_datetime(power_df['utctime'])
power_df.index = power_df["utctime"]

poll_df["ts_utc"] = pandas.to_datetime(poll_df['ts_utc'])
poll_df.index = poll_df["ts_utc"]
```

```python
tol = pandas.Timedelta('50 milliseconds')
tol_2 = pandas.Timedelta('5 seconds')
df = pandas.merge_asof(left=prof_df, right=power_df,right_index=True,left_index=True,direction='nearest',tolerance=tol)
df = pandas.merge_asof(left=df, right=poll_df,right_index=True,left_index=True,direction='nearest',tolerance=tol_2)

pandas.set_option('display.max_columns', None)
pandas.set_option('display.expand_frame_repr', False)

display(df)
```
